```python
import requests
from bs4 import BeautifulSoup


def get_poem_metadata(url: str) -> dict:
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()

        soup = BeautifulSoup(response.text, "html.parser")

        title_tag = soup.find("meta", property="og:title")
        description_tag = soup.find("meta", property="og:description")
        author_tag = soup.find("meta", attrs={"name": "author"})

        metadata = {
            "title": title_tag.get("content") if title_tag else None,
            "description": description_tag.get("content") if description_tag else None,
            "author": author_tag.get("content") if author_tag else None,
            "url": url,
        }

        return metadata

    except requests.RequestException as e:
        print(f"Request failed: {url}")
        return None
 
```

```python
get_poem_metadata('https://www.aldiwan.net/poem24653.html')
```

```python
urls = ['https://www.aldiwan.net/poem24653.html',
        'https://www.aldiwan.net/poem24654.html',
        'https://www.aldiwan.net/poem24655.html']

for url in urls:
    data = get_poem_metadata(url)
    print(data)
```
